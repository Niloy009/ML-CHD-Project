```python
#K fold Models on smote data with feature selection
import numpy as npy
import pandas as pds
import matplotlib.pyplot as mplt
import seaborn as sbn
%matplotlib inline

train_data = pds.read_csv('trainDataSMOTE.csv')
test_data = pds.read_csv('testData.csv')
```

```python
train_dataF = train_data[['sysBP', 'glucose', 'age', 'totChol', 'male', 'diaBP', 'prevalentHyp', 'diabetes', 'BPMeds', 'cigsPerDay', 'TenYearCHD']]
test_dataF =  test_data[['sysBP', 'glucose', 'age', 'totChol', 'male', 'diaBP', 'prevalentHyp', 'diabetes', 'BPMeds', 'cigsPerDay', 'TenYearCHD']]
test_dataF.head(10)
```

```python
train_input = train_dataF.iloc[:,0:10]
train_label = train_dataF.iloc[:,-1]
test_input = test_dataF.iloc[:,0:10]
test_label = test_dataF.iloc[:,-1]
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression

SKF = StratifiedKFold(n_splits = 10)
score_skfLR = []
score_testLR = []
i = 1
for train_index, test_index in SKF.split(train_input, train_label):
    print('{} number KFold from {}'.format(i,SKF.n_splits))
    xtr_inputs, xval_inputs = train_input.loc[train_index], train_input.loc[test_index]
    ytr_label, yval_label = train_label.loc[train_index], train_label.loc[test_index]
    
    logistic_model = LogisticRegression(solver = 'liblinear')
    logistic_model.fit(xtr_inputs, ytr_label)
    accu = logistic_model.score(xval_inputs, yval_label)
    score_skfLR.append(accu)
    print('Validation Fold accuracy : ', accu)
    accu = logistic_model.score(test_input, test_label)
    score_testLR.append(accu)
    print('Test set accuracy : ', accu, '\n\n')
    i+=1
```

```python
print('Validation Fold score : ',score_skfLR,'\n\nMean Validation Score',npy.mean(score_skfLR))
print('\nTest score : ',score_testLR,'\n\nMean Test Score',npy.mean(score_testLR))
```

```python
from sklearn.neighbors import KNeighborsClassifier

score_skfK1 = []
score_skfK5 = []
score_skfK10 = []
score_testK1 = []
score_testK5 = []
score_testK10 = []
i = 1
for train_index, test_index in SKF.split(train_input, train_label):
    print('{} number KFold from {}'.format(i,SKF.n_splits))
    xtr_inputs, xval_inputs = train_input.loc[train_index], train_input.loc[test_index]
    ytr_label, yval_label = train_label.loc[train_index], train_label.loc[test_index]
    
    k1_model = KNeighborsClassifier(n_neighbors = 1, metric = 'chebyshev').fit(xtr_inputs, ytr_label)
    k5_model = KNeighborsClassifier(n_neighbors = 5, metric = 'chebyshev').fit(xtr_inputs, ytr_label)
    k10_model = KNeighborsClassifier(n_neighbors = 10, metric = 'chebyshev').fit(xtr_inputs, ytr_label)

    score_skfK1.append(k1_model.score(xval_inputs, yval_label))
    score_skfK5.append(k5_model.score(xval_inputs, yval_label))
    score_skfK10.append(k10_model.score(xval_inputs, yval_label))
    score_testK1.append(k1_model.score(test_input, test_label))
    score_testK5.append(k5_model.score(test_input, test_label))
    score_testK10.append(k10_model.score(test_input, test_label))
    i+=1
```

```python
print('Validation Folds score \nKNN 1 : ', score_skfK1, '\nKNN2 : ', score_skfK5, '\nKNN 10 : ', score_skfK10)
print('\n\nValidation Mean score \nKNN 1 : ', npy.mean(score_skfK1), '\nKNN 5 : ', npy.mean(score_skfK5), '\nKNN 10 : ', npy.mean(score_skfK10))
print('\n\nTests score \nKNN 1 : ', score_testK1, '\nKNN2 : ', score_testK5, '\nKNN 10 : ', score_testK10)
print('\n\nTests mean score \nKNN 1 : ', npy.mean(score_testK1), '\nKNN 5 : ', npy.mean(score_testK5), '\nKNN 10 : ', npy.mean(score_testK10))
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

vc_model = VotingClassifier(estimators=[('1', LogisticRegression(solver = 'liblinear')), ('2', SVC(kernel = 'linear', probability = True)), ('3', DecisionTreeClassifier(criterion = 'entropy', max_depth = 5))], voting='soft')
```

```python
SKF = StratifiedKFold(n_splits = 10)
score_skfLR = []
score_testLR = []
i = 1
for train_index, test_index in SKF.split(train_input, train_label):
    print('{} number KFold from {}'.format(i,SKF.n_splits))
    xtr_inputs, xval_inputs = train_input.loc[train_index], train_input.loc[test_index]
    ytr_label, yval_label = train_label.loc[train_index], train_label.loc[test_index]
    
    vc_model.fit(xtr_inputs, ytr_label)
    accu = vc_model.score(xval_inputs, yval_label)
    score_skfLR.append(accu)
    print('Validation Fold accuracy : ', accu)
    accu = vc_model.score(test_input, test_label)
    score_testLR.append(accu)
    print('Test set accuracy : ', accu, '\n\n')
    i+=1
```

```python
print('Validation Fold score : ',score_skfLR,'\n\nMean Validation Score',npy.mean(score_skfLR))
print('\nTest score : ',score_testLR,'\n\nMean Test Score',npy.mean(score_testLR))
```
